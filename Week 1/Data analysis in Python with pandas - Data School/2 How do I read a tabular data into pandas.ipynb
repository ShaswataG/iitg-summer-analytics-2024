<h1>3 How do I select a pandas Series from a DataFrame</h1>

```python
# both methods can be used for reading csv file - read_table method with sep argument and read_csv method
# ufo = pd.read_table("http://bit.ly/uforeports", sep=",")
ufo = pd.read_csv("http://bit.ly/uforeports")
```

```python
ufo
```

```python
print(type(ufo["City"]))
ufo["Colors Reported"]
# instead of useing square brackets, we can use dot notation
# all the series are present inside the ufo object as attributes
ufo.City
```

```python
# ufo.Colours Reported
# when the series or column name has space in it, dot notation doesn't work, and we must use bracket notation
ufo["Colors Reported"]
```

```python
# when the series or column name is a keyword, dot notation doesn't work, and we must use bracket notation. For ex: head, shape
```

<h3>Adding new series or column to a dataframe</h3>

```python
# on the left hand side of '=', we must use bracket notation and not dot notation
ufo["Location"] = ufo.City + ", " + ufo.State
```

```python
ufo
```
